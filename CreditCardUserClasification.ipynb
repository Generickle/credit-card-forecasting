```python
from sklearn.preprocessing import StandardScaler, normalize
```

```python
scaler = StandardScaler()
creditcard_df_scaled = scaler.fit_transform(uci_credit_card_data)
```

```python
creditcard_df_scaled.shape
```

```python
creditcard_df_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
scores = []

range_values = range(2, 20)

for i in range_values:
    kmeans = KMeans(n_clusters = i)

    kmeans.fit(creditcard_df_scaled)
    scores.append(kmeans.inertia_)


plt.plot(range_values, scores, 'bx-')
plt.title("Encontrar el número óptimo de Clusters")
plt.xlabel("Clusters")
plt.ylabel("WCSS(k)")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

silhouette_scores = []

# Probamos desde 2 clusters (no tiene sentido calcular silhouette con 1 solo cluster)
range_values = range(2, 20)

for k in range_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(creditcard_df_scaled)
    
    score = silhouette_score(creditcard_df_scaled, labels)
    silhouette_scores.append(score)

# Graficar Silhouette Scores
plt.plot(range_values, silhouette_scores, 'bx-')
plt.title("Silhouette Scores para distintos valores de k")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
silhouette_scores
```

```python
from sklearn.metrics import silhouette_score
```

```python
# Calculate silhouette scores for different cluster numbers
silhouette_scores = []
for i in range(2, 20):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(creditcard_df_scaled)
    silhouette_scores.append(silhouette_score(creditcard_df_scaled, kmeans.labels_))
plt.plot(range(2, 20), silhouette_scores, 'bx-')
plt.title("Silhouette Score vs. Number of Clusters")
plt.xlabel("Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

color = 'tab:blue'
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('WCSS (Elbow Method)', color=color)
ax1.plot(range_values, scores, 'o-', color=color, label='WCSS (Elbow)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:orange'
ax2.set_ylabel('Silhouette Score', color=color)
ax2.plot(range(2, 20), silhouette_scores, 's-', color=color, label='Silhouette Score')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')

plt.title('Elbow Method vs Silhouette Score')
plt.show()
```

```python
from tensorflow.keras.layers import Input, Add, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, AveragePooling2D, MaxPooling2D, Dropout
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.initializers import glorot_uniform
from keras.optimizers import SGD
```

```python
## Dimensionalidad
encoding_dim = 2

## Estructura del dataset, 24 columnas
input_df = Input(shape = (24, ))

# Glorot normal inicializador (Xavier normal initializer) tomar muestras aleatorias de una distribución normal truncada
x = Dense(encoding_dim, activation = 'relu')(input_df)
x = Dense(500, activation = 'relu', kernel_initializer = 'glorot_uniform')(x)
x = Dense(500, activation = 'relu', kernel_initializer = 'glorot_uniform')(x)
x = Dense(2000, activation = 'relu', kernel_initializer = 'glorot_uniform')(x)

encoded = Dense(10, activation = 'relu', kernel_initializer = 'glorot_uniform')(x)

x = Dense(2000, activation = 'relu', kernel_initializer = 'glorot_uniform')(encoded)
x = Dense(500, activation = 'relu', kernel_initializer = 'glorot_uniform')(x)

decoded = Dense(24, kernel_initializer = 'glorot_uniform')(x)

autoencoder = Model(input_df, decoded)

encoder = Model(input_df, encoded)

autoencoder.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
## Estructura resultante para predecir
creditcard_df_scaled.shape
```

```python
## Modelo de la red neuronal
autoencoder.summary()
```

```python
autoencoder.fit(
    creditcard_df_scaled,
    creditcard_df_scaled,
    batch_size = 128,
    epochs = 25,
    verbose = 1
)
```

```python
## Guardamos nuestro modelo con pesos
autoencoder.save_weights('autoencoder.weights.h5')
```

```python
## Realizamos la predicción
pred = encoder.predict(creditcard_df_scaled)
```
